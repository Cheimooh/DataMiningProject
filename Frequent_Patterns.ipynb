```python
import pandas as pd
import numpy as np
import plotly.express as px
import scipy.stats as stats
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
import seaborn as sns
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
# Charger les données depuis un fichier CSV
file_path = "./data/clean_dataset_googleplaystore_cat_str.csv"
df = pd.read_csv(file_path)

# Sélectionner les colonnes pertinentes
selected_columns = ["GenrePrincipal", "GenreSecondaire", "Installs"]
data = df[selected_columns].values.tolist()

# Convertir les données en format adapté pour l'algorithme Apriori
te = TransactionEncoder()
te_ary = te.fit(data).transform(data)
df_encoded = pd.DataFrame(te_ary, columns=te.columns_)

# Appliquer l'algorithme Apriori pour obtenir les itemsets fréquents
supports = apriori(df_encoded, min_support=0.1, max_len=3, use_colnames=True)

# Afficher les itemsets fréquents
print(supports)
supports.head()
```

```python
df = pd.read_csv("data/clean_dataset_googleplaystore_cat_str.csv")
df.head(2)

# Je veux créer un dataframe sur lequel je vais appliquer l'algorithme Apriori. Pour cela je souhaite
# avoir une ligne par application et difféntes colonnes : price, rating, revies, installs.
# Dans la colonne price, je souhaite avoir 1 si l'application est gratuite, 0 sinon.
# Dans la colonne rating, je souhaite avoir 1 si l'application a une note supérieure à 4.5, 0 sinon.
# Dans la colonne reviews, je souhaite avoir 1 si l'application a plus de 100000 reviews, 0 sinon.
# Dans la colonne installs, je souhaite avoir 1 si l'application a plus de 10000000 installs, 0 sinon.
# Code ce dataframe dans la variable df_apriori

df_apriori = df.copy()
df_apriori["Price"] = df_apriori["Price"].apply(lambda x: 1 if x == 0 else 0)
df_apriori["Rating"] = df_apriori["Rating"].apply(lambda x: 1 if x > 4.5 else 0)
df_apriori["Reviews"] = df_apriori["Reviews"].apply(lambda x: 1 if x > 100000 else 0)
# df_apriori["installs"] = df_apriori["Installs"].apply(lambda x: 1 if x > 10000000 else 0)
# Ajouter les colonnes genre principal et genre secondaire avec comme valeur 1 pour l'indice du tableau qui est le bon genre    
df_apriori = pd.concat([df_apriori, pd.get_dummies(df_apriori["GenrePrincipal"])], axis=1)
df_apriori = pd.concat([df_apriori, pd.get_dummies(df_apriori["GenreSecondaire"])], axis=1)
df_apriori = df_apriori[["App", "Price", "Rating", "Reviews", "GenrePrincipal", "GenreSecondaire", "Installs"]]

# Séparer les genres principaux et secondaires en colonnes, chaque genre devient une colonne et le nom de cette colonne est la concaténation
# de son nom et de 1 s'il est dans le genre principal ou 2 s'il est dans le genre secondaire. Comme valeur à l'intérieur de la colonne, on met 1
# si l'application est dans ce genre, 0 sinon.
df_apriori["GenrePrincipal"] = df_apriori["GenrePrincipal"].astype(str) + "1"
df_apriori["GenreSecondaire"] = df_apriori["GenreSecondaire"].astype(str) + "2"
#Séparer les genres principaux et secondaires en colonnes de type int64
df_apriori = pd.concat([df_apriori, pd.get_dummies(df_apriori["GenrePrincipal"])], axis=1)
df_apriori = pd.concat([df_apriori, pd.get_dummies(df_apriori["GenreSecondaire"])], axis=1)

# Si une colonne à le type uint8, on la convertit en int64
for column in df_apriori.columns:
    if df_apriori[column].dtype == "uint8":
        df_apriori[column] = df_apriori[column].astype("int64")

#On vire les colonnes genre principal et genre secondaire
# df_apriori = df_apriori.drop(columns=["GenrePrincipal", "GenreSecondaire"])

df_apriori.info()

# Enlever les lignes où installs n'est pas à 1
# df_apriori = df_apriori[df_apriori["Installs"] == 1]

df_apriori.head(2)
```

```python
# Exclure des colonnes
excluded_columns = ["App", "Installs", "Reviews", "GenrePrincipal", "GenreSecondaire"]
pivoted = df_apriori.pivot_table(index='App', columns='Installs', values=[col for col in df_apriori.columns if col not in excluded_columns], aggfunc='mean')

# Créer la liste des colonnes sélectionnées
selected_columns = [col for col in df_apriori.columns if col not in excluded_columns]

# Afficher la liste des colonnes sélectionnées
print(selected_columns)
pivoted = pivoted.fillna(0)
pivoted = pivoted.astype(int)

pivoted.head()
```

```python
supports = apriori(pivoted, min_support=0.005,max_len=3, use_colnames=True)
```

```python
supports.sort_values(by="support", ascending=False).head()
```

```python
supports["len"] = supports["itemsets"].apply(lambda x: len(x))
pd.set_option("display.max_colwidth", None)
supports[supports["len"]==2].sort_values(by="support", ascending=False).head(10)
```

```python
# compute association rules
rules = association_rules(supports, min_threshold=0.5)
rules.head(2)
```
